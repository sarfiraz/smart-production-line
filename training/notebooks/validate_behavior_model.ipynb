```python
mon = IOHealthMonitor(response_timeout_cycles=3)

for i in range(4):
    faults = mon.update(
        {"I_1":0,"I_2":0,"I_3":1,"I_4":0},
        {"PWM_1":80,"PWM_2":0,"PWM_3":0,"PWM_4":0}
    )
    print(i+1, faults)
```

```python
mon = IOHealthMonitor(stuck_cycles=3)

tests = [
    {"I_1":0,"I_2":1,"I_3":1,"I_4":0},
    {"I_1":0,"I_2":0,"I_3":0,"I_4":0},
    {"I_1":0,"I_2":1,"I_3":1,"I_4":0},
    {"I_1":0,"I_2":0,"I_3":1,"I_4":0},
]

for i,ins in enumerate(tests,1):
    faults = mon.update(ins, {"PWM_1":0,"PWM_2":0,"PWM_3":0,"PWM_4":0})
    print(i, faults)
```
